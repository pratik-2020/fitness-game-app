```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
X = np.arange(-6, 6, 0.1)
Y = np.zeros(len(X))
```

#Binary Step Function

```python
def binary_step(x):
    if x<0:
        return 0
    else:
        return 1
for i in range(len(X)):
  Y[i] = binary_step(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.style.use('dark_background')
#plt.grid()
plt.ylim(-3, 3)
plt.plot(X, Y)
plt.show()
```

#Linear Function

```python
def linear(x):
    return 4*x

for i in range(len(X)):
  Y[i] = linear(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.plot(X, Y)
plt.grid()
plt.show()
```

#Sigmoid Funtion

```python
def sigmoid(x):
    z = (1/(1 + np.exp(-x)))
    return z
#sigmoid derivative --> sigmoid(x)*(1-sigmoid(x))
for i in range(len(X)):
  Y[i] = sigmoid(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-6, 6)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#Tanh

```python
def tanh(x):
    z = (2/(1 + np.exp(-2*x))) -1 # 2sigmoid(2x)-1
    return z
for i in range(len(X)):
  Y[i] = tanh(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-6, 6)
plt.ylim(-2, 2)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#ReLU

```python
def relu(x):
    if x<0:
        return 0
    else:
        return x

for i in range(len(X)):
  Y[i] = relu(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#Leaky ReLU

```python
def leaky_relu(x):
    if x<0:
        return 0.01*x
    else:
        return x

for i in range(len(X)):
  Y[i] = leaky_relu(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#Parameterised ReLU


```python
def pReLU(x,a):
    if x<0:
        return a*x
    else:
        return x
for i in range(len(X)):
  Y[i] = pReLU(X[i], 0.5)

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()

```

#Exponential Linear Unit(ELU)

```python
def elu(x, a):
    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x

for i in range(len(X)):
  Y[i] = elu(X[i], 0.5)

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#Swish

```python
def swish(x):
    return x/(1-np.exp(-x)) 

for i in range(len(X)):
  Y[i] = swish(X[i])

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
 
#depict illustration
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.plot(X, Y)
plt.style.use('dark_background')
plt.show()
```

#Softmax

```python
def softmax(x): #here x is an array/list of probabilities
    z = np.exp(x)
    z_ = z/z.sum()
    return z_ 
X_ = np.array([0.2, 0.4, 0.3, 0.1])
Y_= np.zeros(len(X_))
Y_ = softmax(X_)
print(Y)
plt.plot(X_, Y_)
plt.style.use('dark_background')
plt.show()
```
